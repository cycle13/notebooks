```python
import pyart
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
%matplotlib inline
```

```python
filename = '/data/radar/IKA 2014021207/201402120710_IKA.PPI3_A.raw'
```

```python
radar = pyart.io.read(filename)
```

```python
print radar.fields.keys()
```

```python
print radar.latitude, radar.longitude
```

```python
font = {'size' : 16}
rc('font', **font)
f = plt.figure(figsize=[15,8])
myd = pyart.graph.RadarMapDisplay(radar)
myd.plot_ppi_map('reflectivity', 0, vmin=-20, vmax=24, resolution = 'h',
                 lat_lines = np.arange(59,63,.5),lon_lines = np.arange(20,26,1) )

myd.plot_range_ring(200.0)
myd.plot_range_ring(150.0, line_style = 'k--')
myd.plot_range_ring(100.0)
myd.plot_range_ring(50.0, line_style = 'k--')
myd.plot_line_xy(np.array([-250000.0, 250000.0]), np.array([0.0,0.0]), line_style = 'k-')
myd.plot_line_xy(np.array([0.0,0.0]),np.array([-250000.0, 250000.0]), line_style = 'k-')


myd.plot_point(radar.longitude['data'], radar.latitude['data'])
myd.plot_point( 24.287847,61.843536, label_text = 'AMF2', label_offset=[.05,.05], symbol = 'bo' )
ax1 = plt.gca()
x0,x1 = ax1.get_xlim()
y0,y1 = ax1.get_ylim()
ax1.set_aspect((x1-x0)/(y1-y0) - .05)

#plt.text(xp, yp, 'foo')
```
