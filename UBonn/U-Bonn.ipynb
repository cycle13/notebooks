```python
#first we do some imports and check the version of Py-ART for consistency
import pyart
from matplotlib import pyplot as plt
import netCDF4
import numpy as np
from copy import deepcopy
%matplotlib inline
print pyart.__version__
```

Use <a href ="https://code.google.com/p/netcdf4-python/"> Python-netCDF4 </a> to read in the processed University of Bonn X-Band data (we need to post the data on line eventually for openness). 

```python
mydataset = netCDF4.Dataset('/data/2011-06-22--11:21zbiasout.nc')
```

Interrogate the data and the metadata

```python
print(mydataset.variables.keys())
for var in mydataset.ncattrs():
    print('Key :', var, ' Value: ', getattr(mydataset, var))

for dim in mydataset.dimensions.keys():
    print('Dimension:', dim, ' Value: ', len(mydataset.dimensions[dim]))
```

The next step is to get the data into the Py-ART Common Data Model 
(here is a <a href = "http://nbviewer.ipython.org/urls/raw.githubusercontent.com/ARM-DOE/notebooks/master/ASR_PI_2014/Investigating%20the%20Py-ART%20Radar%20Object.ipynb?create=1"> Notebook </a> explaining the data model). This involves setting up a translation matrix between the names in the netCDF file and more Py-ART friendly names plus populating metadata etc..

Populate an empty radar object

```python
field_translation = {'Kdpglatt+' : 'glatt_diff_phase', 
                     'PHIdpglatt': 'glatt_dp_phase_shift',  
                     'Kdpplot' : 'plot_diff_phase',
                     'PHIdpplot' : 'plot_dp_phase_shift',
                     'Zdrplot' : 'plot_diff_reflectivity',
                     'delta' : 'delta_phase_shift',
                     'Zh' : 'reflectivity_horizontal', 
                     'Zv' : 'reflectivity_vertical', 
                     'Zhfil' : 'corrected_reflectivity_horizontal', 
                     'Kdp' : 'diff_phase', 
                     'PHIdp' : 'dp_phase_shift',
                     'Zdr' : 'diff_reflectivity',
                     'Uh' : 'raw_reflectivity_horizontal', 
                     'Uv' : 'raw_reflectivity_vertical', 
                     'Vh' : 'mean_doppler_velocity', 
                     'Vv' : 'vertical_mean_doppler_velocity', 
                     'Rhv' : 'copol_coeff'}


ngates = len(mydataset.dimensions['range'])
nrays = len(mydataset.dimensions['azi'])
myradar = pyart.testing.make_empty_ppi_radar(ngates, nrays, 1)
for field in field_translation.keys(): 
    #print(field)
    field_dict = {'data' : mydataset.variables[field][:],
                  'units' : 'Unk',
                  'long_name' : mydataset.variables[field].long_name}
    myradar.add_field(field_translation[field], field_dict)   
myradar.range['data'] = np.linspace(0,ngates*100.0, ngates)
```

Lets see how some of the data looks! 

```python
fields_to_plot = ['dp_phase_shift', 'reflectivity_horizontal']
ranges = [(-92, 90), (0, 52)]
display = pyart.graph.RadarMapDisplay(myradar)

nplots = len(fields_to_plot)
plt.figure(figsize=[7 * nplots, 4])
for plot_num in xrange(nplots):
    field = fields_to_plot[plot_num]
    vmin, vmax = ranges[plot_num]
    plt.subplot(1, nplots, plot_num + 1)
    display.plot_ppi_map(field, 0, vmin=vmin, vmax=vmax, title_flag=False)
    display.plot_range_rings([50,100])
    #display.set_limits(ylim=[-100, 100], xlim=[-100,100])
```

Now we are going to run the stardard ARM processing on the phase data as showing in (Giangrande et al 2013). This requires us to set the FZL (going to set it to 2km) and the system phase and fill in a Normalized Coherent Power field which identifies first trip echoes. 

```python
#hard set the FZL
FZL = 2000.0
#hard set system phase
sysphase = -93.0
#create false NCP field 

ncp_dict = {'data' : np.ones(myradar.fields['copol_coeff']['data'].shape),
            'standard_name' : 'signal_quality_index',
            'long_name' : 'normalized_coherent_power'}

myradar.add_field('norm_coherent_power', ncp_dict)
```

Now fun the LP Processing code

```python
phidp, kdp = pyart.correct.phase_proc_lp(myradar, 0.0, debug = True, fzl = FZL, 
                                         overide_sys_phase = True,
                                         sys_phase = sysphase, 
                                         refl_field = 'reflectivity_horizontal',
                                         phidp_field = 'dp_phase_shift',
                                         rhv_field = 'copol_coeff',
                                         ncp_field = 'norm_coherent_power',
                                         self_const=60000.0, LP_solver='cylp')
myradar.fields['proc_dp_phase_shift'] = phidp
myradar.fields['recalculated_diff_phase'] = kdp
```

This returns and adds a $\phi_{dp}$ and $K_{dp}$ field. Lets Plot!

```python
display = pyart.graph.RadarDisplay(myradar)
fig = plt.figure(figsize=[10, 5])
ax = fig.add_subplot(111)
ray_num = 185

# filtered phidp and unfolded phidp
display.plot_ray('proc_dp_phase_shift', ray_num, format_str='b-',
                 axislabels_flag=False, title_flag=False, ax=ax)

display.plot_ray('unfolded_differential_phase', ray_num, format_str='g-',
                 axislabels_flag=False, title_flag=False, ax=ax)

# set labels
ax.set_ylim(0, 140)
ax.set_ylabel('Differential phase shift (degrees)')
ax.set_xlabel('Range (km)')

# plot KDP and reflectivity on second axis
ax2 = ax.twinx()
display.plot_ray('recalculated_diff_phase', ray_num, format_str='r-',
                 axislabels_flag=False, title_flag=False, ax=ax2)

display.plot_ray('glatt_diff_phase', ray_num, format_str='r--',
                 axislabels_flag=False, title_flag=False, ax=ax2)

display.plot_ray('diff_phase', ray_num, format_str='r:',
                 axislabels_flag=False, title_flag=False, ax=ax2)

display.plot_ray('scaled_reflectivity', ray_num, format_str='b-',
                 axislabels_flag=False, title_flag=False, ax=ax2)

# decorate
ax2.yaxis.grid(color='gray', linestyle='dashed')
ax.legend(display.plots,
          ["Filtered phiDP", "Unfolded phiDP", 'LP KDP','Glat KDP','Raw? KDP', 'Z/10.0'],
          loc='upper left')
ax.set_xlim(0, 40)
ax2.set_ylim(-1, 15)
```
