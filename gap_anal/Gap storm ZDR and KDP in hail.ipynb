Polarimetric analysis of the CP2 Data from the time of the Gap Storm.

```python
import pyart
from matplotlib import pyplot as plt
import numpy as np
from matplotlib import rc
%matplotlib inline
```

```python
filename = '/data/radar/cp2/063604.mdv'
radar = pyart.io.read(filename)
```

```python
print radar.fields.keys()
```

```python
gap_loc = [ 152.9531, -27.4439]
```

```python
font = {'size' : 16}
rc('font', **font)
f = plt.figure(figsize=[15,8])
myd = pyart.graph.RadarMapDisplay(radar)
myd.plot_ppi_map('reflectivity', -1, vmin=-8, vmax=72, 
                 min_lon=152.5, max_lon=153.15, min_lat=-27.9, max_lat=-27.4,
                 lon_lines = np.arange(150, 154, .2), 
                 lat_lines = np.arange(-35, -27, .1), resolution = 'h',
                 auto_range=False)
myd.plot_range_ring(25. * 1000.0, line_style = 'k-')
myd.plot_range_ring(50. * 1000.0, line_style = 'k--')

myd.plot_line_xy([-50000.0, 50000.0], [0.0,0.0], line_style = 'k-')
myd.plot_line_xy([0.0,0.0],[-50000.0, 50000.0], line_style = 'k-')

myd.plot_point(radar.longitude['data'], radar.latitude['data'])
myd.plot_point( gap_loc[0], gap_loc[1], symbol = 'ro', label_text='The Gap',  
               label_offset=[0.01, 0.01] )

ax1 = plt.gca()
x0,x1 = ax1.get_xlim()
y0,y1 = ax1.get_ylim()
ax1.set_aspect((x1-x0)/(y1-y0) - .15)

#plt.text(xp, yp, 'foo'
```
