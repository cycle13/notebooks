```python
def bbox2ij(lon,lat,bbox=[-160., -155., 18., 23.]):
    """Return indices for i,j that will completely cover the specified bounding box.     
    i0,i1,j0,j1 = bbox2ij(lon,lat,bbox)
    lon,lat = 2D arrays that are the target of the subset
    bbox = list containing the bounding box: [lon_min, lon_max, lat_min, lat_max]

    Example
    -------  
    >>> i0,i1,j0,j1 = bbox2ij(lon_rho,[-71, -63., 39., 46])
    >>> h_subset = nc.variables['h'][j0:j1,i0:i1]       
    """
    bbox=np.array(bbox)
    mypath=np.array([bbox[[0,1,1,0]],bbox[[2,2,3,3]]]).T
    p = path.Path(mypath)
    points = np.vstack((lon.flatten(),lat.flatten())).T   
    n,m = np.shape(lon)
    inside = p.contains_points(points).reshape((n,m))
    ii,jj = np.meshgrid(xrange(m),xrange(n))
    return min(ii[inside]),max(ii[inside]),min(jj[inside]),max(jj[inside])
```

First lets put together our procedures for segmentation. First is a procedure to break down the image into segements based on an input array of rain rates (R). The code then returns the total area at which this rain rate is equal to or exceeded (A(R)), the number or objects (N(R)) and the mean rain rate across the objects ($\mathrm{R_m(R)}$). The second bit uses a [Watersheding algorithm](http://en.wikipedia.org/wiki/Watershed_%28image_processing%29)  to segment the rainfall field. A watershed algorithm starts flow from various seed points in the image and fills until the basin meets at certain points and then identifies this as a region. Initial attempts seeding at all local maxima (minima as we use -1.0*R) produced too complicated a segmentation so we used a smoother in our seed point determination in order to avoid small scale structure and noise. 

```python
def area_anal(pixel_area, rr_x, rain_rates):
    A_rainrate=np.zeros(rr_x.shape)
    N_rainrate=np.zeros(rr_x.shape)
    Rm_rainrate=np.zeros(rr_x.shape)
    my_shape=rain_rates.shape
    for i in range(len(rr_x)):
        b_fld=np.zeros(my_shape)
        b_fld[rain_rates > rr_x[i]]=1.0
        regions, N_rainrate[i] = ndimage.label(b_fld)
        try:
            A_rainrate[i]=len(np.where(regions > 0.5)[0])*pixel_area
            Rm_rainrate[i]=rain_rates[np.where(regions > 0.5)].mean()
        except IndexError:
            A_rainrate[i]=0.0
            Rm_rainrate[i]=0.0
    return N_rainrate, A_rainrate, Rm_rainrate
        
def dt_to_dict(dt, **kwargs):
    pref = kwargs.get('pref', '')
    return dict([(pref+key, getattr(dt, key)) for key in
                ['year', 'month', 'day', 'hour', 'minute', 'second']])
```

$f(R)=ae^{-\frac{(x-b)^2}{2c^2}}$
$=a\frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{1}{2}\bigl(\frac{x-\mu}{\sigma^2}\bigr)^2}$

```python
def my_func(xdata,a1,a2,center1, center2, sigma1, sigma2):
    r2p=np.sqrt(2.0*np.pi)
    f1=a1*(1.0/(sigma1*r2p))*np.exp(-0.5*((xdata-center1)/(sigma1**2))**2)
    f2=a2*(1.0/(sigma2*r2p))*np.exp(-0.5*((xdata-center2)/(sigma2**2))**2)
    return f1+f2
    
ig=[100,10, 5,50, np.sqrt(5),np.sqrt(10)]
params=np.zeros([N_all.shape[0], len(ig)])-1.0
success=np.zeros([N_all.shape[0]])
for i in range(N_all.shape[0]):
    try:
        popt, pcov = scipy.optimize.curve_fit(my_func, rr_x, N_all[i,:], ig)
        params[i,:]=popt
        success[i]=1
        #ig=popt
    except RuntimeError:
        success[i]=0
print success
suc=np.where(success==1.0)[0]
fail=np.where(success!=1.0)[0]
for fail_index in fail:
    cs=np.array((suc-fail_index)**2).argsort()[0]
    closest_success=suc[cs]
    ig=params[closest_success,:]
    try:
        popt, pcov = scipy.optimize.curve_fit(my_func, rr_x, N_all[fail_index,:], ig)
        params[fail_index,:]=popt
        success[fail_index]=1
    except RuntimeError:
        success[fail_index]=0
suc=np.where(success==1.0)[0]
fail=np.where(success!=1.0)[0]
for fail_index in fail[::-1]:
    cs=np.array((suc-fail_index)**2).argsort()[0]
    closest_success=suc[cs]
    ig=params[closest_success,:]
    try:
        popt, pcov = scipy.optimize.curve_fit(my_func, rr_x, N_all[fail_index,:], ig)
        params[fail_index,:]=popt
        success[fail_index]=1
    except RuntimeError:
        success[fail_index]=0
print success
```
