```python
import pyart
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
fn ='/data/234845.mdv'
myradar = pyart.io.read(fn)
```

```python
nyq = myradar.instrument_parameters['nyquist_velocity']['data'][0]
my_display = pyart.graph.RadarDisplay(myradar)
tilt = 1
f = plt.figure(figsize=[15,5])
plt.subplot(1,2,1)
my_display.plot_ppi('reflectivity', tilt, vmin=-6, vmax=64.0)
plt.xlim([-50,0])
plt.ylim([-100,-50])
plt.subplot(1,2,2)
my_display.plot_ppi('differential_reflectivity', tilt, vmin=-1, vmax=6)
plt.xlim([-50,0])
plt.ylim([-100,-50])
```

```python
nyq = myradar.instrument_parameters['nyquist_velocity']['data'][0]
my_display = pyart.graph.RadarDisplay(myradar)
tilt = 1
f = plt.figure(figsize=[15,5])
plt.subplot(1,2,1)
my_display.plot_ppi('reflectivity', tilt, vmin=-6, vmax=64.0)
plt.xlim([-50,0])
plt.ylim([-100,-50])
plt.subplot(1,2,2)
my_display.plot_ppi('velocity', tilt, vmin=-nyq, vmax=nyq)
plt.xlim([-50,0])
plt.ylim([-100,-50])
```

```python
snr = pyart.retrieve.calculate_snr_from_reflectivity(myradar)
myradar.add_field('SNR', snr, replace_existing = True)
gatefilter = pyart.correct.GateFilter(myradar)
gatefilter.exclude_below('SNR', 10) #Non-Sig

corr_vel = pyart.correct.dealias_region_based(
    myradar, vel_field='velocity', keep_original=False, 
skip_along_ray=100, skip_between_rays=100, gatefilter = gatefilter)
myradar.add_field('corrected_mean_doppler_velocity', corr_vel, True)
```

```python
nyq = myradar.instrument_parameters['nyquist_velocity']['data'][0]
my_display = pyart.graph.RadarDisplay(myradar)
tilt = 1
f = plt.figure(figsize=[15,5])
plt.subplot(1,2,1)
my_display.plot_ppi('reflectivity', tilt, vmin=-6, vmax=64.0)
plt.xlim([-50,0])
plt.ylim([-100,-50])
plt.subplot(1,2,2)
my_display.plot_ppi('corrected_mean_doppler_velocity', tilt, vmin=-2.0*nyq, vmax=2.0*nyq)
plt.xlim([-50,0])
plt.ylim([-100,-50])
```
