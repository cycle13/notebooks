First lets put together our procedures for segmentation. First is a procedure to break down the image into segements based on an input array of rain rates (R). The code then returns the total area at which this rain rate is equal to or exceeded (A(R)), the number or objects (N(R)) and the mean rain rate across the objects ($\mathrm{R_m(R)}$). The second bit uses a [Watersheding algorithm](http://en.wikipedia.org/wiki/Watershed_%28image_processing%29)  to segment the rainfall field. A watershed algorithm starts flow from various seed points in the image and fills until the basin meets at certain points and then identifies this as a region. Initial attempts seeding at all local maxima (minima as we use -1.0*R) produced too complicated a segmentation so we used a smoother in our seed point determination in order to avoid small scale structure and noise. 

```python
def area_anal(pixel_area, rr_x, rain_rates):
    A_rainrate=np.zeros(rr_x.shape)
    N_rainrate=np.zeros(rr_x.shape)
    Rm_rainrate=np.zeros(rr_x.shape)
    my_shape=rain_rates.shape
    for i in range(len(rr_x)):
        b_fld=np.zeros(my_shape)
        b_fld[rain_rates > rr_x[i]]=1.0
        regions, N_rainrate[i] = ndimage.label(b_fld)
        try:
            A_rainrate[i]=len(np.where(regions > 0.5)[0])*pixel_area
            Rm_rainrate[i]=rain_rates[np.where(regions > 0.5)].mean()
        except IndexError:
            A_rainrate[i]=0.0
            Rm_rainrate[i]=0.0
    return N_rainrate, A_rainrate, Rm_rainrate
        
        
        


def water_anal(grid,rr, label_min=5.0, **kwargs):
    b_fld=np.zeros(grid.shape)
    b_fld[grid > rr]=1.0
    image=deepcopy(grid.data)
    image[image < 0.0]=0.0
    regions, numobjects = ndimage.label(b_fld)
    if 'presmth' in kwargs:
        image_rep=deepcopy(image)
        rainsmth=ndimage.filters.gaussian_filter(image_rep, kwargs['presmth'])
        local_maxi = is_local_maximum(rainsmth, regions)
    else:
        local_maxi = is_local_maximum( image, regions)
    markers = ndimage.label(local_maxi)[0]
    labels = watershed(-image, markers, mask=image>label_min)
    #healing single pixel finds
    labels_healed=deepcopy(labels)
    nob=0
    for i in range(labels_healed.max()):
        map_points=np.where(labels_healed == i)
        if len(map_points[0])==1:
            #print i, "huzah!", len(map_points[0])
            labels_healed[map_points]=labels_healed[map_points[0]-1,map_points[1]]
    new_array=numpy.zeros(labels.shape, dtype=int)        
    for i in range(labels_healed.max()):
        map_points=np.where(labels_healed == i+1)
        if len(map_points[0])>1:
            #print i, len(map_points[0])
            nob=nob+1
            #print i, "huzah!", len(map_points[0])
            new_array[map_points]=nob 
    return new_array

def labels_to_verts(xx,yy,mylabels):
    objects=[]
    numobjects=mylabels.max()
    for i in range(numobjects):
        this_object=np.zeros(mylabels.shape)
        this_object[mylabels==i]=1.0
        cs = plt.contour(xx, yy,this_object, [0.5])
        p = cs.collections[0].get_paths()[0]
        v = p.vertices
        x = v[:,0]
        y = v[:,1]
        objects.append((x,y))
    return objects

def dt_to_dict(dt, **kwargs):
    pref = kwargs.get('pref', '')
    return dict([(pref+key, getattr(dt, key)) for key in
                ['year', 'month', 'day', 'hour', 'minute', 'second']])
```

Pull the rain rate info from each grid. Using the watershed technique determine a meshgrid of labels (1 to n) and vertecies for each time step and store in a list.

```python
rain=[mygrids[i].fields['rain_rate_A']['data'][0,0,:,:] for i in range(len(mygrids))]
labels=[]
for thisgrid in rain:
    labels.append(water_anal(thisgrid, 1.0, label_min=1.0, presmth=5.0))
cells=[]
for label in labels:
    cells.append(labels_to_verts(mygrids[0].axes['x_disp']['data'],mygrids[0].axes['y_disp']['data'], label ))
```
