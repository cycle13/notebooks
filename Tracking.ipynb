```python
import matplotlib.pyplot as plt
import sys
import os
pyart_dir=os.environ.get('PYART_DIR',os.environ['HOME']+'/python')
sys.path.append(pyart_dir)
from pyart.io import  radar, grid
from pyart.graph import rayplot, radar_display
from pyart.io import nc_utils
import netCDF4
from copy import deepcopy
import numpy as np
from scipy import ndimage
from skimage.morphology import watershed, is_local_maximum 
from skimage.segmentation import random_walker
from matplotlib import animation
```

```python
def area_anal(pixel_area, rr_x, rain_rates):
    A_rainrate=np.zeros(rr_x.shape)
    N_rainrate=np.zeros(rr_x.shape)
    Rm_rainrate=np.zeros(rr_x.shape)
    my_shape=rain_rates.shape
    for i in range(len(rr_x)):
        b_fld=np.zeros(my_shape)
        b_fld[rain_rates > rr_x[i]]=1.0
        regions, N_rainrate[i] = ndimage.label(b_fld)
        try:
            A_rainrate[i]=len(np.where(regions > 0.5)[0])*pixel_area
            Rm_rainrate[i]=rain_rates[np.where(regions > 0.5)].mean()
        except IndexError:
            A_rainrate[i]=0.0
            Rm_rainrate[i]=0.0
    return N_rainrate, A_rainrate, Rm_rainrate
        
        
        

def water_anal(grid,rr):
    b_fld=np.zeros(grid.shape)
    b_fld[grid > rr]=1.0
    image=deepcopy(grid.data)
    image[image < 1.0]=0.0
    regions, numobjects = ndimage.label(b_fld)
    local_maxi = is_local_maximum( image, regions)
    markers = ndimage.label(local_maxi)[0]
    labels = watershed(-image, markers, mask=image>5.0)
    #healing single pixel finds
    labels_healed=deepcopy(labels)
    nob=0
    for i in range(labels_healed.max()):
        map_points=np.where(labels_healed == i)
        if len(map_points[0])==1:
            #print i, "huzah!", len(map_points[0])
            labels_healed[map_points]=labels_healed[map_points[0]-1,map_points[1]]
    new_array=numpy.zeros(labels.shape, dtype=int)        
    for i in range(labels_healed.max()):
        map_points=np.where(labels_healed == i+1)
        if len(map_points[0])>1:
            #print i, len(map_points[0])
            nob=nob+1
            #print i, "huzah!", len(map_points[0])
            new_array[map_points]=nob 
    return new_array

def labels_to_verts(xx,yy,mylabels):
    objects=[]
    numobjects=mylabels.max()
    for i in range(numobjects):
        this_object=np.zeros(mylabels.shape)
        this_object[mylabels==i]=1.0
        cs = plt.contour(xx, yy,this_object, [0.5])
        p = cs.collections[0].get_paths()[0]
        v = p.vertices
        x = v[:,0]
        y = v[:,1]
        objects.append((x,y))
    return objects

def dt_to_dict(dt, **kwargs):
    pref = kwargs.get('pref', '')
    return dict([(pref+key, getattr(dt, key)) for key in
                ['year', 'month', 'day', 'hour', 'minute', 'second']])
```

```python
ind='/data/qpe/20110520/'
files=os.listdir(ind)
want_files=[]
for file in files:
    if ('20110520.07' in file) or ('20110520.08' in file) or ('20110520.09' in file) or ('20110520.10' in file):
        want_files.append(ind+file)

want_files.sort()
```

```python
mygrids=[]
for filen in want_files:
    print filen
    myncf=netCDF4.Dataset(filen)
    mygrids.append(grid.pyGrid(myncf))
    myncf.close()
```

```python
rain=[mygrids[i].fields['rain_rate_A']['data'][0,0,:,:] for i in range(len(mygrids))]
labels=[]
for thisgrid in rain:
    labels.append(water_anal(thisgrid, 1.0))
cells=[]
for label in labels:
    cells.append(labels_to_verts(mygrids[0].axes['x_disp']['data'],mygrids[0].axes['y_disp']['data'], label ))
```

```python
def update(n, labels, verts, xyt, ax): 
    ax.cla()
    ax.pcolormesh(xyt[0], xyt[1], labels[n], vmin=0, vmax=80)
    for obj in verts[n]:
        ax.plot(obj[0]/1000.0, obj[1]/1000.0, 'k-')
    ax.set_xlabel('E/W distance from CF')
    ax.set_ylabel('N/S distance from CF')
    tit=" Watershed segmentation,  %(year)04d-%(month)02d-%(day)02d %(hour)02d:%(minute)02d:%(second)02d" %dt_to_dict(xyt[2][n])
    ax.set_title(tit)
    
t=[netCDF4.num2date(mygrids[i].axes['time']['data'][0], 
                    mygrids[i].axes['time']['units'],
                    'gregorian') for i in range(len(mygrids))]

xyt=[mygrids[0].axes['x_disp']['data']/1000.0,mygrids[0].axes['y_disp']['data']/1000.0, t]

fig = plt.figure(figsize=[8,5])
ax = fig.add_axes([0.1, 0.1, .9, .8])
pc=ax.pcolormesh(xyt[0], xyt[1], labels[0], vmin=0, vmax=80)
cb=colorbar(mappable=pc)
cb.set_label(r'Rainfall rate (mmhr$^{-1}$)')
anim = animation.FuncAnimation(fig, update, frames=len(labels), blit=True, fargs=(rain, cells, xyt, ax))
anim.save('/data/water_track.mp4', fps=5, bitrate=100, dpi=200);
```
