```python
import netCDF4
import os
import sys
from matplotlib import pyplot as plt
import matplotlib.dates as mdates
import numpy as np
%matplotlib inline
```

```python
all_files = os.listdir('/data/swats/')
swats = []
for fil in all_files:
    if '.cdf' in fil:
        swats.append(fil)
swats.sort()
```

```python
datenums = mdates.date2num(netCDF4.num2date(myncf.variables['time_offset'][:], myncf.variables['time_offset'].units))
fig = plt.figure(figsize = [15,7])

for i in range(len(myncf.variables['depth'][:])):
    plt.plot(datenums, myncf.variables['soilwatpot_W'][:,i])


hours   = mdates.HourLocator()
tFmt    = mdates.DateFormatter('%H:%MZ')
plt.gca().xaxis_date()
plt.gca().xaxis.set_major_locator(hours)
plt.gca().xaxis.set_major_formatter(tFmt)
plt.xlabel('Time')
plt.ylabel(myncf.variables['watcont_W'].long_name + ' ('+ myncf.variables['watcont_W'].units+')')
```

```python
soil = np.empty([0, 8])
watcont = np.empty([0, 8])
time = np.empty(0)
for i in range(len(swats)):
    myncf = netCDF4.Dataset('/data/swats/'+ swats[i])
    #print myncf.variables['soilwatpot_W'].shape
    #print i*6,(i+1)*6
    soil = np.append(soil, myncf.variables['soilwatpot_W'][:], axis = 0)
    watcont = np.append(watcont, myncf.variables['watcont_W'][:], axis = 0)
    time = np.append(time, mdates.date2num(netCDF4.num2date(myncf.variables['time_offset'][:], 
                                                        myncf.variables['time_offset'].units)))
    myncf.close()
```
