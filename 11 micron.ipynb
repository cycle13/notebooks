```python
import netCDF4
from matplotlib import pyplot as plt
import pyproj
import numpy as np
from mpl_toolkits.basemap import Basemap

%matplotlib inline
```

```python
data=netCDF4.Dataset('http://motherlode.ucar.edu:8080/thredds/dodsC/satellite/IR/WEST-CONUS_4km/current/WEST-CONUS_4km_IR_20131115_2245.gini')
```

```python
print data.variables.keys()
print data.variables['x'].units
print data.ncattrs()
print data.Conventions
print data.geospatial_lat_min
print data.geospatial_lon_min
print data.summary

print data.variables['IR'].units
print data.variables['IR'][:].min()
print data.variables['IR'][:].max()
print data.variables['IR'].ncattrs()
print data.variables['IR'].scale_factor

print data.variables['IR'].long_name
```

```python
vals=data.variables['IR'][:]*-1.0
swap_me = np.where(vals < 0)
vals[swap_me] = vals[swap_me] + 128.0*2.0

print vals.shape
xg, yg = np.meshgrid(data.variables['x'][:]*1000.0, data.variables['y'][:]*1000.0)

pnyc = pyproj.Proj(proj = 'lcc',datum = 'NAD83', 
                   lat_0 = data.geospatial_lat_max,
                   lon_0 = data.geospatial_lon_max,
                   x_0 = data.variables['x'][0]*1000.0,
                   y_0 = data.variables['y'][0]*1000.0)
lon, lat = pnyc(xg, yg, inverse = True)
```

```python
min_lon = lon.min(); max_lon = lon.max()
min_lat = lat.min(); max_lat = lat.max()

box=[-90, -80, 38, 46]

m = Basemap(llcrnrlon=box[0],llcrnrlat=box[2],
            urcrnrlon=box[1],urcrnrlat=box[3],
            projection='mill',area_thresh=10000,
            resolution='h')

x, y = m(lon, lat)
# create figure.
fig=plt.figure(figsize=(10,9))
ax = fig.add_axes([0.01,0.05,0.9,0.80]) 
m.drawparallels(np.linspace(38,46, 9) ,labels=[1,1,0,0])
m.drawmeridians(np.linspace(-90, -80,11),labels=[0,0,0,1]) 
m.pcolormesh(x, y , vals[-1,:], cmap=plt.get_cmap('binary'), vmin=150, vmax=170)
xr, yr = m([box[0], box[1]], [box[2], box[3]])
plt.xlim(xr)
plt.ylim(yr)

m.drawcoastlines(linewidth=1.25)
m.drawstates()
plt.colorbar()
```
