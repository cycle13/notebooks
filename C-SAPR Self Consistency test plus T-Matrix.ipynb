```python
# For testing variable aspect ratio
def drop_ar(D_eq):
    if D_eq < 0.7:
        return 1.0;
    elif D_eq < 1.5:
        return 1.173 - 0.5165*D_eq + 0.4698*D_eq**2 - 0.1317*D_eq**3 - \
            8.5e-3*D_eq**4
    else:
        return 1.065 - 6.25e-2*D_eq - 3.99e-3*D_eq**2 + 7.66e-4*D_eq**3 - \
            4.095e-5*D_eq**4 
```

```python
n_d0=15
n_Nw=15
D0=np.linspace(0.2, 4.5, n_d0)
Nw=np.linspace(1e2, 10e4, n_Nw)
zdr=np.zeros([n_d0,n_Nw])
z=np.zeros([n_d0,n_Nw])
kdp=np.zeros([n_d0,n_Nw])
A=np.zeros([n_d0,n_Nw])
for i in range(n_d0):
    for j in range(n_Nw):
        tm = TMatrixPSD(lam=tmatrix_aux.wl_C, m=refractive.m_w_10C[tmatrix_aux.wl_C])
        tm.psd_eps_func = lambda D: 1.0/drop_ar(D)
        tm.D_max = 10.0
        tm.or_pdf = orientation.gaussian_pdf(20.0)
        tm.orient = orientation.orient_averaged_fixed
        tm.geometries = (tmatrix_aux.geom_horiz_back, tmatrix_aux.geom_horiz_forw)
        tm.init_scatter_table()
        tm.psd = GammaPSD(D0=D0[i], Nw=Nw[j], mu=4)
        zdr[i,j]=radar.Zdr(tm)
        z[i,j]=radar.refl(tm)
        tm.set_geometry(tmatrix_aux.geom_horiz_forw)
        kdp[i,j]=radar.Kdp(tm)
        A[i,j]=radar.Ai(tm)

```

```python
form=r'N$_w$=%(ee)2.1e m$^{-3}$'
labels=[form %{'ee':this_nw} for this_nw in Nw]
```

```python
plots=[]
fig = plt.figure(figsize=[12,7])
pcm = plt.pcolormesh(sc_yedges, sc_xedges, np.log10(sc_H+0.01), vmin=0.0)
for i in range(len(labels)):
    plots.append(plt.plot(kdp[:,i],10.0*np.log10(z[:,i]), label=labels[i]))

plt.xlabel('Specific differential Phase ($\mathrm{\circ{}/km}$)')
plt.ylabel('Eq. Reflectivity Factor (dBZ)');cb=plt.colorbar()
plt.gca().axes.set_ylim(refl_range)
plt.gca().axes.set_xlim(kdp_range)
cb.set_label(r'log$_{10}$($\mathrm{n_{gates})}$')
title=plt.title('Self consistency test of $\mathrm{Z_e}$ and $\mathrm{K_{DP}}$ ')
plt.legend(loc=4, framealpha=0.3, fontsize=10)

```
